13.**What is the total monthly rent of your unit?**: 31 missing fields, Updated with median operation and perfomed the below
data cleaning operations listed as follows on this column

1. Defined the column name for easy reference.
2. checks for missing values before starting the cleaning process.
3. converted all entries in the column to numeric values, coercing any non-numeric values to NaN (not a number).
4. calculated the median rent of the column after coercing non-numeric values to ensure that the median is based only on valid numeric entries.
5. filled in missing values with the calculated median rent.
6. Removed any non-numeric characters like dollar signs and commas from the entries and converts the values to floats.
7. Replaced entries that contain "4 months" with NaN since this is not a valid rent value.
8. Replaced any rent values above $3000 with NaN, as you've determined that such values are not realistic for your dataset.
9. Recalculates the median rent after removing unrealistic values and fills in NaNs with this new median value.
10.Finally, printed out the descriptive statistics of the cleaned column to check the results of the cleaning process.

```python
# Define the column name
rent_column_name = "What is the total monthly rent of your unit?"

# Check Missing Values Before Cleaning
missing_values_before = questionnaire_responses[rent_column_name].isnull().sum()
print(f"Missing values in '{rent_column_name}' before cleaning:", missing_values_before)

# Ensure that all non-numeric data is set to NaN
questionnaire_responses[rent_column_name] = pd.to_numeric(questionnaire_responses[rent_column_name], errors='coerce')

# Now calculate the median of the column after cleaning
median_rent = questionnaire_responses[rent_column_name].median()

# Fill in the missing values with the median
questionnaire_responses[rent_column_name].fillna(median_rent, inplace=True)

# Check Missing Values After Filling
missing_values_after = questionnaire_responses[rent_column_name].isnull().sum()
print(f"Missing values in '{rent_column_name}' after filling:", missing_values_after)

# Remove non-numeric characters and convert to float
questionnaire_responses['What is the total monthly rent of your unit?'] = (
    questionnaire_responses['What is the total monthly rent of your unit?']
    .replace('[\$,]', '', regex=True)  # Remove dollar signs and commas
    .replace('4 months', np.nan, regex=False)  # Replace '4 months' with NaN
    .replace(r'^\s*$', np.nan, regex=True)  # Replace empty strings with NaN
    .astype(float)  # Convert to float
)

# Replace values that are unrealistically high with NaN
questionnaire_responses.loc[
    questionnaire_responses['What is the total monthly rent of your unit?'] > 3000,
    'What is the total monthly rent of your unit?'
] = np.nan

# Replace NaN values with the median rent
median_rent = questionnaire_responses['What is the total monthly rent of your unit?'].median()
questionnaire_responses['What is the total monthly rent of your unit?'].fillna(median_rent, inplace=True)

# Now, you can check the results
print(questionnaire_responses['What is the total monthly rent of your unit?'].describe())
```

16.**If yes, what is your monthly rental insurance?**: replace missing and non-numeric values in the monthly rental insurance column with zero, under the assumption that these responses indicate no payment. It also sets values outside a reasonable range (0 to 5000) to zero and

The data in column shows a wide variety of entries including numeric values, strings that represent numeric values (like '13 dollars'), strings that denote the absence of a cost or knowledge thereof (like 'No', 'Don’t know exactly', 'Dont know', 'na'), and even some nonsensical entries (like very long numbers that are likely input errors).

Below code implement will handle such data accordingly:

Convert strings with dollar amounts to numeric values.
Treat phrases indicating absence of cost or lack of knowledge as missing or zero, depending on the context (e.g., 'No' and 'na' can be treated as 0).
Handle nonsensical numeric entries (likely errors) by setting a reasonable upper limit for insurance costs, above which values are set to NaN or 0.
Impute missing values with a reasonable statistic, such as the median of the non-missing, non-erroneous values.

```python
# Define the column name for monthly rental insurance
rental_insurance_column_name = "If yes, what is your monthly rental insurance?"

# Function to clean and convert insurance values
def clean_insurance_value(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        value = value.lower()
        if any(non_cost_phrase in value for non_cost_phrase in ['no', 'don’t know', 'dont know', 'na', 'zero', '-', ' ']):
            return 0.0
        # Remove non-numeric characters and convert to float
        value = ''.join(filter(str.isdigit, value))
        return float(value) if value else 0.0
    return float(value)

# Apply the function to the rental insurance column
questionnaire_responses[rental_insurance_column_name] = questionnaire_responses[rental_insurance_column_name].apply(clean_insurance_value)

# Set an upper limit for insurance costs
upper_limit_insurance = 5000  # Set this based on what you deem to be a reasonable upper limit
questionnaire_responses.loc[questionnaire_responses[rental_insurance_column_name] > upper_limit_insurance, rental_insurance_column_name] = np.nan

# Impute missing values with median after cleaning
median_insurance = questionnaire_responses[rental_insurance_column_name].median()
questionnaire_responses[rental_insurance_column_name].fillna(median_insurance, inplace=True)

# Check missing values after cleaning and imputation
missing_values_after_cleaning = questionnaire_responses[rental_insurance_column_name].isnull().sum()
print(f"Missing values after cleaning and imputation in '{rental_insurance_column_name}':", missing_values_after_cleaning)
```

17.**What is the approximate total amount of monthly bills (electricity, parking, etc.,) not included in your rent?**: The data in this column contains both numeric and non-numeric values, as well as potentially erroneous data. The presence of strings, such as '180 dollars', '200 per 5 members', and nonsensical numeric values, suggests that we need to clean this data before proceeding with the median imputation.

Here's an approach to clean the data and handle missing values:

Convert all the entries to a numeric format where possible, setting non-numeric entries to NaN.
Remove any nonsensical numeric values that are likely input errors.
After cleaning, calculate the median of the remaining numeric values.
Impute missing values with the median.

```python
# Define the column name
monthly_bills_column_name = "What is the approximate total amount of monthly bills (electricity, parking, etc.,) not included in your rent?"

# Check Missing Values Before Cleaning
missing_values_before_monthly_bills = questionnaire_responses[monthly_bills_column_name].isnull().sum()
print(f"Missing values in '{monthly_bills_column_name}' before cleaning:", missing_values_before_monthly_bills)

# Function to clean and convert values to numeric
def clean_monthly_bills(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        # Remove any non-digit characters and convert to float
        cleaned_value = ''.join(filter(str.isdigit, value))
        return float(cleaned_value) if cleaned_value else np.nan
    return value

# Apply the cleaning function to the monthly bills column
questionnaire_responses[monthly_bills_column_name] = questionnaire_responses[monthly_bills_column_name].apply(clean_monthly_bills)

# Remove any nonsensical values by setting a reasonable upper limit for monthly bills
upper_limit_monthly_bills = 5000  # This should be based on domain knowledge or additional data analysis
questionnaire_responses.loc[questionnaire_responses[monthly_bills_column_name] > upper_limit_monthly_bills, monthly_bills_column_name] = np.nan

# Calculate the median of the column after cleaning
median_monthly_bills = questionnaire_responses[monthly_bills_column_name].median()

# Fill in the missing values with the median
questionnaire_responses[monthly_bills_column_name].fillna(median_monthly_bills, inplace=True)

# Check Missing Values After Filling
missing_values_after_monthly_bills = questionnaire_responses[monthly_bills_column_name].isnull().sum()
print(f"Missing values in '{monthly_bills_column_name}' after filling:", missing_values_after_monthly_bills)
```
